```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('datasets\\Family Income and Expenditure.csv')

df
```

<h3>CHANGING DATA TYPES</h3>

```python
# --- Convert Text Columns (Objects) to String ---
df['Region'] = df['Region'].astype('string')
df['Main Source of Income'] = df['Main Source of Income'].astype('string')
df['Household Head Sex'] = df['Household Head Sex'].astype('string')
df['Household Head Marital Status'] = df['Household Head Marital Status'].astype('string')
df['Household Head Highest Grade Completed'] = df['Household Head Highest Grade Completed'].astype('string')
df['Household Head Job or Business Indicator'] = df['Household Head Job or Business Indicator'].astype('string')
df['Household Head Occupation'] = df['Household Head Occupation'].astype('string')
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].astype('string')
df['Type of Household'] = df['Type of Household'].astype('string')
df['Type of Building/House'] = df['Type of Building/House'].astype('string')
df['Type of Roof'] = df['Type of Roof'].astype('string')
df['Type of Walls'] = df['Type of Walls'].astype('string')
df['Tenure Status'] = df['Tenure Status'].astype('string')
df['Toilet Facilities'] = df['Toilet Facilities'].astype('string')
df['Main Source of Water Supply'] = df['Main Source of Water Supply'].astype('string')

# --- Convert Flags (Integers) to Boolean ---
df['Agricultural Household indicator'] = df['Agricultural Household indicator'].astype('bool')
df['Electricity'] = df['Electricity'].astype('bool')

# --- Check the result ---
df.info()
```

<h3>CHECKING AND FIXING NULL VALUES</h3>

```python
df.isnull().sum()
```

```python
df['Household Head Occupation'] = df['Household Head Occupation'].fillna('Unemployed')
df['Household Head Class of Worker'] = df['Household Head Class of Worker'].fillna('Unemployed')
toilet_mode = df['Toilet Facilities'].mode()[0]
df['Toilet Facilities'] = df['Toilet Facilities'].fillna(toilet_mode)
df.isnull().sum()
```

<h1>DESCRIPTIVE ANALYSIS</h1>

<h2>wants Young vs old
<h2/>

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# SETTINGS
sns.set_style("white")
plt.rcParams['figure.figsize'] = (12, 10) # Made the chart slightly taller to fit new items

# ---------------------------------------------------------
# 1. DATA PREPARATION
# ---------------------------------------------------------
young_mask = df['Household Head Age'] < 35
old_mask = df['Household Head Age'] >= 60

# UPDATED: Added Communication and Miscellaneous
wants_cols = {
    'Dining Out': 'Restaurant and hotels Expenditure',
    'Alcohol': 'Alcoholic Beverages Expenditure',
    'Tobacco': 'Tobacco Expenditure',
    'Fashion/Clothing': 'Clothing, Footwear and Other Wear Expenditure',
    'Parties/Occasions': 'Special Occasions Expenditure',
    'Comm/Internet': 'Communication Expenditure',             # NEW
    'Personal Care/Misc': 'Miscellaneous Goods and Services Expenditure' # NEW
}

comparison_data = []

for label, col in wants_cols.items():
    young_avg = df.loc[young_mask, col].mean()
    old_avg = df.loc[old_mask, col].mean()
    comparison_data.append({'Category': label, 'Young': young_avg, 'Old': old_avg})

df_wants = pd.DataFrame(comparison_data)
df_wants = df_wants.sort_values(by='Young', ascending=True)

# ---------------------------------------------------------
# 2. VISUALIZATION: BUTTERFLY CHART
# ---------------------------------------------------------
fig, ax = plt.subplots()

y_pos = range(len(df_wants))
bar_height = 0.6

# Plot Bars
ax.barh(y_pos, -df_wants['Young'], height=bar_height, align='center', color='#3498db', label='Young Generation (<35)')
ax.barh(y_pos, df_wants['Old'], height=bar_height, align='center', color='#e74c3c', label='Older Generation (60+)')

# Styling
ax.set_yticks(y_pos)
ax.set_yticklabels(df_wants['Category'], fontsize=12)

# Fix X-axis labels
xticks = ax.get_xticks()
ax.set_xticklabels([f'P{abs(x):,.0f}' for x in xticks])

# Add Value Labels
for i, (young, old) in enumerate(zip(df_wants['Young'], df_wants['Old'])):
    ax.text(-young - 100, i, f'P{young:,.0f}', ha='right', va='center', color='#3498db', fontweight='bold')
    ax.text(old + 100, i, f'P{old:,.0f}', ha='left', va='center', color='#e74c3c', fontweight='bold')

ax.axvline(0, color='grey', linewidth=1)

plt.title('The Generational Wants Clash: Digital vs. Traditional', fontsize=16, pad=20)
plt.xlabel('Average Annual Expenditure (PHP)', fontsize=12)
plt.legend(loc='lower right')
sns.despine(left=True, bottom=True)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# 1. CREATE DECADE BINS
# We group ages into 20s, 30s, 40s...
bins = [19, 29, 39, 49, 59, 69, 100]
labels = ['20s', '30s', '40s', '50s', '60s', '70+']
df['Decade'] = pd.cut(df['Household Head Age'], bins=bins, labels=labels)

# 2. CALCULATE AVERAGE EXPENSE PER DECADE
decade_expenses = df.groupby('Decade')['Total Expenses'].mean()

# 3. PLOT
plt.figure(figsize=(10, 6))
colors = sns.color_palette("Blues", len(decade_expenses)) # Light to Dark Blue

bars = plt.bar(decade_expenses.index, decade_expenses.values, color=colors, edgecolor='black', alpha=0.8)

# 4. ADD LABELS AND "JUMP" PERCENTAGE
for i in range(len(bars)):
    height = bars[i].get_height()
    plt.text(bars[i].get_x() + bars[i].get_width()/2, height + 5000,
             f'P{height:,.0f}', ha='center', fontweight='bold')

    # Calculate the jump from the previous decade (Prediction Insight)
    if i > 0:
        prev_height = bars[i-1].get_height()
        pct_change = ((height - prev_height) / prev_height) * 100
        direction = "+" if pct_change > 0 else ""

        # Put the percentage in the middle of the bar
        plt.text(bars[i].get_x() + bars[i].get_width()/2, height/2,
                 f'{direction}{pct_change:.0f}%',
                 ha='center', color='white', fontweight='bold', fontsize=12)

plt.title('The 10-Year Outlook: How Expenses Change by Decade', fontsize=16)
plt.ylabel('Average Annual Expense', fontsize=12)
plt.xlabel('Life Stage', fontsize=12)
plt.show()
```

<h2>Fiesta
<h2/>

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1. Create Synthetic Data (Similar to your dataset)
np.random.seed(42)
n_samples = 1000
regions = ['NCR', 'I - Ilocos', 'IV-A CALABARZON', 'VII - Central Visayas',
           'XI - Davao', 'V - Bicol', 'III - Central Luzon', 'ARMM']
data = {
    'Region': np.random.choice(regions, n_samples),
    'Total Household Income': np.random.lognormal(11, 0.8, n_samples),
    'Special Occasions Expenditure': np.random.lognormal(9, 1.2, n_samples)
}
df = pd.DataFrame(data)

# 2. Prepare Data for Charts
plt.style.use("seaborn-v0_8-whitegrid")
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 5))

# --- CHART A: The "Lollipop Chart" (Regional Ranking) ---
# Group by Region and get Median spending (Median is better to ignore outliers)
regional_stats = df.groupby('Region')['Special Occasions Expenditure'].median().sort_values(ascending=True)

# Create the Lollipop stems
my_range = range(1, len(regional_stats.index) + 1)
ax1.hlines(y=my_range, xmin=0, xmax=regional_stats, color='skyblue', alpha=0.8, linewidth=3)
ax1.plot(regional_stats, my_range, "o", markersize=15, color='#007acc', alpha=0.9)


# Labels and Titles
ax1.set_yticks(my_range)
ax1.set_yticklabels(regional_stats.index, fontsize=11)
ax1.set_title("Which Region Spends the Most on Occasions?\n(Median Expenditure)", fontsize=14, fontweight='bold', color='#333333')
ax1.set_xlabel("Median Spending (Php)", fontsize=11)
ax1.grid(axis='y', alpha=0.3)

# --- CHART B: The "Trend Line" (Income vs Spending) ---
# Create Income "Bins" (groups) to smooth out the noise
# We divide income into 10 quantiles (deciles)
df['Income_Bin'] = pd.qcut(df['Total Household Income'], q=10)
# Calculate mean spending for each income bin
trend_data = df.groupby('Income_Bin')['Special Occasions Expenditure'].mean().reset_index()

# Get the mid-point of each bin for plotting the X-axis
trend_data['Bin_Center'] = trend_data['Income_Bin'].apply(lambda x: x.mid)

# Plot the Area Chart
ax2.fill_between(range(len(trend_data)), trend_data['Special Occasions Expenditure'], color="skyblue", alpha=0.3)
ax2.plot(range(len(trend_data)), trend_data['Special Occasions Expenditure'], color="#007acc", linewidth=3, marker='o')

# Formatting labels (Making the X-axis readable)
# We just show "Low Income" to "High Income" to keep it simple for infographics
ax2.set_xticks([0, 9])
ax2.set_xticklabels(["Poorest 10%", "Richest 10%"], fontsize=12, fontweight='bold')
ax2.set_title("The Fiesta Celebration Effect\n(How Spending Jumps as Income Rises)", fontsize=14, fontweight='bold', color='#333333')
ax2.set_ylabel("Average Occasion Spending (Php)", fontsize=12)
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

<h2>male vs women savings
<h2/>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Set the Style to WhiteGrid (Clean, professional look)
sns.set_style("whitegrid")

# Create a figure with 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# Define colors (Keeping your specific Pink/Blue hex codes)
palette_colors = {'Female': '#ff824d', 'Male': '#3470a3'}

# -------------------------------------------------------
# CHART 1: Violin Plot (Distribution)
# -------------------------------------------------------
# The 'inner' parameter adds lines for the quartiles inside the violin
sns.violinplot(x='Household Head Sex', y='Savings Rate', data=df,
               palette=palette_colors, ax=ax1, inner='quartile', split=True)

ax1.set_title("Distribution of Savings Behavior\n(Are they consistent?)", fontsize=14, fontweight='bold', color='#333333')
ax1.set_ylabel("Savings Rate (%)", fontsize=12)
ax1.set_xlabel("") # Remove x-label as the ticks are self-explanatory

# -------------------------------------------------------
# CHART 2: Scatter Plot with Trend Lines (Context)
# -------------------------------------------------------
# Scatter plot to show individual data points
sns.scatterplot(x='Total Household Income', y='Savings Rate', hue='Household Head Sex',
                data=df, palette=palette_colors, alpha=0.6, s=60, ax=ax2)

# Add Trend Lines (The regression lines)
# This reveals the saving trend as income rises
sns.regplot(x='Total Household Income', y='Savings Rate', data=df[df['Household Head Sex']=='Male'],
            scatter=False, color='#22a6b3', ax=ax2, label='Male Trend')
sns.regplot(x='Total Household Income', y='Savings Rate', data=df[df['Household Head Sex']=='Female'],
            scatter=False, color='#e056fd', ax=ax2, label='Female Trend', line_kws={'linestyle':'--'})

# Log scale is crucial for income data to see the pattern clearly
ax2.set_xscale('log')

ax2.set_title("Savings Rate vs. Income Level\n(Who is thriftier at the same salary?)", fontsize=14, fontweight='bold', color='#333333')
ax2.set_xlabel("Total Household Income (Log Scale)", fontsize=12)
ax2.set_ylabel("Savings Rate (%)", fontsize=12)
ax2.legend()
plt.tight_layout()
plt.show()
```
